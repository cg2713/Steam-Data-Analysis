# Pandas setup/import

```python
import pandas as pd
```

#File paths, loading, display

```python
# Paste file paths into following variables

# First Iteration 
page = 3
file1 = f'../../../data_collection/data/{page}/steam_app_data_{page}.csv'        # 1st file to merge
file2 = f'../../../data_collection/data/{page}/steamspy_data_{page}.csv'        # 2nd file to merge
#file3 = ''        # 3rd file to merge

dest1 = f'../{page}/steamdata{page}.csv'        # 1st destination file, post merge
```

```python
# Load the files

dataframe1 = pd.read_csv(file1)
dataframe2 = pd.read_csv(file2)
```

## rearanging dataframe

```python
appidrelocal = dataframe1.pop('steam_appid')
dataframe1.insert(0, 'steam_appid', appidrelocal)
```

```python
dataframe1.head()
```

```python
dataframe2.rename(columns = {'appid':'steam_appid'},inplace=True)
dataframe2.head()
```

# Merge 1

```python
# This creates a new merged dataset, used to cross check information

# Usage
  # on: Which column is consistent between the two dataframes? 
    # Organizations and Funding Rounds, use uuid 
    # Funding Rounds and Investments, use funding_round_uuid
  # how: Think of a merge in terms of venn diagram. 
    # 'inner' is middle section. 
    # 'outer' is everything. 
    # Those two types are only important ones.

CV_Merge1 = pd.merge(dataframe1, dataframe2,
             on='steam_appid',       
             how='inner')      

CV_Merge1.head(20)  # Check if our data lines up.
```

```python
# Save the dataframe to a csv file
CV_Merge1.to_csv(dest1)
```
