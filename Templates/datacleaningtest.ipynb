```python
prices = df.loc[:,"price_overview"]
prices
```

```python
print(len(prices), end=' ')
keyValue = re.compile("(?P<key>'final'): (?P<value>\d*)")
chP = keyValue.findall(prices[0])
print(chP[0][1])
#prices
```

```python
df.loc[0,'price_overview']
```

```python
#https://www.youtube.com/watch?v=43MvCJulF1g use this to get rid of settingwithcopy warning incase it shows up

for inx, i in enumerate(tqdm(prices, desc="Cleaning Computation: ", unit="Rows")):
    if str(i).find('nan') != 0:
        chP = keyValue.findall(i)
        #df.loc[1,:'price_overview']
        df.loc[inx,'price_overview'] = decimal.Decimal(int(chP[0][1])) / 100
        #prices[inx] = decimal.Decimal(int(chP[0][1])) / 100 # warning occures
        #df.loc[prices, inx] = decimal.Decimal(int(chP[0][1])) / 100



#for inx, i in enumerate(prices):
#    if str(i).find('nan') != 0:
#        chP = keyValue.findall(i)
#        prices[inx] = decimal.Decimal(int(chP[0][1])) / 100
```
